# Rotations in the Bloch Sphere

```python
import qiskit
import numpy as np
import matplotlib.pyplot as plt
import warnings
from qiskit.tools.visualization import plot_bloch_multivector, plot_histogram
warnings.filterwarnings('ignore')
```

$$\newcommand{\ket}[1]{\left|{#1}\right\rangle}$$
$$\newcommand{\bra}[1]{\left\langle{#1}\right|}$$
$$\newcommand{\braket}[2]{\left\langle{#1}\middle|{#2}\right\rangle}$$

```python
simulator=qiskit.Aer.get_backend('statevector_simulator')
def show_state(circuit):
    return qiskit.execute(circuit, backend=simulator).result().get_statevector()
```

## Some Bloch Vectors

### X axis

```python
circuit = qiskit.QuantumCircuit(1,1)
circuit.ry(np.pi/2,0)
plot_bloch_multivector(show_state(circuit))
```

### What are the components of the state vector? (not the Block vector)

Hint: the state is 50% $\ket{0}$ and 50% $\ket{1}$ and there is no phase shift.


Answer:

$\ket{\Psi} = r_0\ket{0}+r_1e^{i\phi}\ket{1}$

$r_0=r_1=\sqrt{\frac{1}{2}}$

$\phi=0$

<img src="spherical-coordinates.png" style="width:200px;"/>

$\ket{\Psi}=\sqrt{\frac{1}{2}}\ket{0}+e^{0i}\sqrt{\frac{1}{2}}\ket{1}$

$\ket{\Psi}=\sqrt{\frac{1}{2}}(1,1)=\sqrt{\frac{1}{2}}(\ket{0}+\ket{1})$

### What are the components of a Bloch vector along the X axis?

Answer: the Bloch vector is $(1,0,0)$

## Y Axis

What are the components of a Bloch vector along the Y axis?

```python
circuit = qiskit.QuantumCircuit(1,1)
circuit.rx(-np.pi/2,0)
plot_bloch_multivector(show_state(circuit))
```

Bloch vector = $(0,1,0)$

## Z Axis

### What are the components of a Bloch vector along the z axis?

```python
circuit = qiskit.QuantumCircuit(1,1)
circuit.h(0)
plot_bloch_multivector(show_state(circuit))
```

The Bloch vector is $(0,0,1)$

## Rotation in the Bloch sphere

## Hadamard Gate

This gate mixes $\ket{0}$ and $\ket{1}$ equally.

$H=\sqrt{\frac{1}{2}}\begin{bmatrix}1&1\\1&-1\end{bmatrix}$

### If we start with $\ket{0}$ and apply a Hadamard gate, where do we end up?

Hint: The final state is 50% up and 50% down, and the phase is zero.

<img src="spherical-coordinates.png" style="width:300px;"/>

```python
circuit = qiskit.QuantumCircuit(1,1)
circuit.h(0)
plot_bloch_multivector(show_state(circuit))
```

### What are the components of this Bloch vector?

Answer: $(1,0,0)$

## If we state with $\ket{1}$ and apply a Hadamard gate, where do we end up?

Hint: the final state is 50% up and 50% down. There is a 180 degree phase shift.

<img src="spherical-coordinates.png" style="width:300px;"/>

```python
circuit = qiskit.QuantumCircuit(1,1)
circuit.x(0)
circuit.h(0)
plot_bloch_multivector(show_state(circuit))
```

### What are the components of this Bloch vector?

Answer: $(-1,0,0)$
