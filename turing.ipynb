1. Mark first cell on the tape
2. Go right until blank is found
2. write 'a' on the tape
3. go left until a marked cell is found
4. move right pme step
5. If $ is present, accept
6. Else mark mark the cell and go to step 2

```python
states = ['start', 'accept', 'reject', 'seeking_left','seeking_righta','seeking_rightb',\
          'moved_right']
tape_alphabet='ab $x'
input_alphabet='ab$'
tape='aba$                 '
for c in list(tape):
    assert c in list(tape_alphabet)
def action(state, char):
    assert state in states
    assert char in tape_alphabet
    if state == 'start':
        new_char = 'x'
        move = 'R'
        new_state = 'seeking_righta'
        if char == 'b':
            new_state = 'seeking_rightb'
    elif state == 'seeking_righta':
        if char == ' ':
            move = 'L'
            new_char = 'a'
            new_state = 'seeking_left'
        else: #'a or b'
            move = 'R'
            new_char = char
            new_state = 'seeking_righta'
    elif state == 'seeking_rightb':
        if char == ' ':
            move = 'L'
            new_char = 'b'
            new_state = 'seeking_left'
        else: #'a or b'
            move = 'R'
            new_char = char
            new_state = 'seeking_rightb'
    elif state == 'seeking_left':
        if char == 'x': #found
            new_char = 'x'
            move = 'R'
            new_state = 'moved_right'
        else: #not yet
            new_char = char
            move = 'L'
            new_state = 'seeking_left'
    elif state == 'moved_right':
        if char == '$':
            new_char = char
            move = 'L'
            new_state = 'accept'
        else:
            new_char = 'x'
            move = 'R'
            new_state = 'seeking_righta'
            if char == 'b':
                new_state = 'seeking_rightb'
    assert new_char in tape_alphabet
    assert new_state in states
    return move, new_char, new_state
head = 0
state = 'start'
while state != 'accept' and state != 'reject':
    print(f'state: {state} head {head} tape: {tape}')
    move, new_char, new_state = action(state, tape[head])
    print(new_char)
    tape1 = list(tape)
    tape1[head]=new_char
    tape = ''.join(tape1)
    if move == 'L':
        head -= 1
        if head<0:
            head=0
    else:
        head += 1
    state = new_state
print(f'state: {state} head {head} tape: {tape}')
```
